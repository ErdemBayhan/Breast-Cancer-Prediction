Winsorizing the Dataset according to outliers

```python
from scipy.stats.mstats import winsorize

df_winsorized = df.copy() #To keep the original dataset

numerical_features = df.select_dtypes(include=['float64', 'int64']).drop(columns=['diagnosis'], errors='ignore')

for col in numerical_features:
    if col in df_winsorized.columns and pd.api.types.is_numeric_dtype(df_winsorized[col]):
          if col != 'diagnosis':
                df_winsorized[col] = df_winsorized[col].clip(lower=df_winsorized[col].quantile(0.05), upper=df_winsorized[col].quantile(0.95) )

cols_to_compare = ['area_mean', 'compactness_mean', 'area_worst'] #Comparing to dataset after winsorizing
print("\n--- Original Dataset Statistics ---")
print(df[cols_to_compare].describe())
print("\n--- After winsorizing Dataset Statistics ---")
print(df_winsorized[cols_to_compare].describe())
```

Assinging the 'diagnosis' column as the target variable. As seen the above, df_winsorized dataset assigned.

```python
# Assign the 'diagnosis' column as the target variable (y)
y = df_winsorized['diagnosis']

# Assign all other columns as features (X)
X = df_winsorized.drop('diagnosis', axis=1)

print("\nFirst 5 rows of target variable (y):")
print(y.head())
print("\nFirst 5 rows of features (X):")
display(X.head())
```

Using a StandartScaler to numerical features share a similar range.

```python
from sklearn.preprocessing import StandardScaler
import numpy as np

num_features_for_scaling = X.select_dtypes(include=np.number).columns
scaler = StandardScaler()
X[num_features_for_scaling] = scaler.fit_transform(X[num_features_for_scaling])
X.head()
```

Spilit to Dataset

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

print(f"Training set shape (X_train): {X_train.shape}")
print(f"Testing set shape (X_test): {X_test.shape}")
print(f"Training target shape (y_train): {y_train.shape}")
print(f"Testing target shape (y_test): {y_test.shape}")

```

```python
X_train.head()
```

MODELS TRAINING

```python
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LogisticRegression


#SVM
# Convert column names to strings
X_train.columns = X_train.columns.astype(str)

# Initialize the SVM classifier
svm_classifier = SVC(random_state=42)

# Fit the classifier to the training data
svm_classifier.fit(X_train, y_train)

#DECISION TREE

# Initialize the Decision Tree classifier
dt_classifier = DecisionTreeClassifier(random_state=42)

# Fit the classifier to the training data
dt_classifier.fit(X_train, y_train)

# Predict the target variable for the test set
y_pred_dt = dt_classifier.predict(X_test)

#RANDOM FOREST

# Initialize the Random Forest classifier
rf_classifier = RandomForestClassifier(random_state=42)

# Fit the classifier to the training data
rf_classifier.fit(X_train, y_train)

# Predict the target variable for the test set
y_pred_rf = rf_classifier.predict(X_test)

#KNN

# Initialize the KNN classifier
knn_classifier = KNeighborsClassifier() #KNN doesn't have a random_state parameter

# Fit the classifier to the training data
knn_classifier.fit(X_train, y_train)

# Predict the target variable for the test set
y_pred_knn = knn_classifier.predict(X_test)

#LOGISTIC REGRESSION

# Initialize the Logistic Regression classifier
logreg_classifier = LogisticRegression(random_state=42)

# Fit the classifier to the training data
logreg_classifier.fit(X_train, y_train)

# Predict the target variable for the test set
y_pred_logreg = logreg_classifier.predict(X_test)
```

```python
print("--- Default Hyperparameters ---")

print("\nLogistic Regression:")
print(logreg_classifier.get_params())

print("\nSupport Vector Machine (SVC):")
print(svm_classifier.get_params())

print("\nDecision Tree Classifier:")
print(dt_classifier.get_params())

print("\nRandom Forest Classifier:")
print(rf_classifier.get_params())

print("\nK-Nearest Neighbors Classifier:")
print(knn_classifier.get_params())
```

MODELS EVALUATION AND VÄ°SUALIZATION

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score, confusion_matrix, classification_report
import matplotlib.pyplot as plt
import seaborn as sns

#SVM

# Convert column names to strings
X_test.columns = X_test.columns.astype(str)

# Predict the target variable for the test set
y_pred = svm_classifier.predict(X_test)

# Calculate evaluation metrics
accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred)
recall = recall_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred)
auc_roc = roc_auc_score(y_test, y_pred)

print(f"Accuracy: {accuracy}")
print(f"Precision: {precision}")
print(f"Recall: {recall}")
print(f"F1-score: {f1}")
print(f"AUC-ROC: {auc_roc}")

# Generate confusion matrix
cm = confusion_matrix(y_test, y_pred)
plt.figure(figsize=(6, 4))
sns.heatmap(cm, annot=True, fmt="d", cmap="Blues")
plt.xlabel("Predicted")
plt.ylabel("True")
plt.title("Confusion Matrix(SVM)")
plt.show()

# Generate classification report
print(classification_report(y_test, y_pred))

#DECISION TREE

# Calculate evaluation metrics
accuracy_dt = accuracy_score(y_test, y_pred_dt)
precision_dt = precision_score(y_test, y_pred_dt)
recall_dt = recall_score(y_test, y_pred_dt)
f1_dt = f1_score(y_test, y_pred_dt)
auc_roc_dt = roc_auc_score(y_test, y_pred_dt)

print(f"Accuracy (Decision Tree): {accuracy_dt}")
print(f"Precision (Decision Tree): {precision_dt}")
print(f"Recall (Decision Tree): {recall_dt}")
print(f"F1-score (Decision Tree): {f1_dt}")
print(f"AUC-ROC (Decision Tree): {auc_roc_dt}")

# Generate confusion matrix
cm_dt = confusion_matrix(y_test, y_pred_dt)
plt.figure(figsize=(6, 4))
sns.heatmap(cm_dt, annot=True, fmt="d", cmap="Blues")
plt.xlabel("Predicted")
plt.ylabel("True")
plt.title("Confusion Matrix (Decision Tree)")
plt.show()

# Generate classification report
print(classification_report(y_test, y_pred_dt))

#RANDOM FOREST

# Calculate evaluation metrics
accuracy_rf = accuracy_score(y_test, y_pred_rf)
precision_rf = precision_score(y_test, y_pred_rf)
recall_rf = recall_score(y_test, y_pred_rf)
f1_rf = f1_score(y_test, y_pred_rf)
auc_roc_rf = roc_auc_score(y_test, y_pred_rf)

print(f"Accuracy (Random Forest): {accuracy_rf}")
print(f"Precision (Random Forest): {precision_rf}")
print(f"Recall (Random Forest): {recall_rf}")
print(f"F1-score (Random Forest): {f1_rf}")
print(f"AUC-ROC (Random Forest): {auc_roc_rf}")

# Generate confusion matrix
cm_rf = confusion_matrix(y_test, y_pred_rf)
plt.figure(figsize=(6, 4))
sns.heatmap(cm_rf, annot=True, fmt="d", cmap="Blues")
plt.xlabel("Predicted")
plt.ylabel("True")
plt.title("Confusion Matrix (Random Forest)")
plt.show()

# Generate classification report
print(classification_report(y_test, y_pred_rf))

#KNN

# Calculate evaluation metrics
accuracy_knn = accuracy_score(y_test, y_pred_knn)
precision_knn = precision_score(y_test, y_pred_knn)
recall_knn = recall_score(y_test, y_pred_knn)
f1_knn = f1_score(y_test, y_pred_knn)
auc_roc_knn = roc_auc_score(y_test, y_pred_knn)

print(f"Accuracy (KNN): {accuracy_knn}")
print(f"Precision (KNN): {precision_knn}")
print(f"Recall (KNN): {recall_knn}")
print(f"F1-score (KNN): {f1_knn}")
print(f"AUC-ROC (KNN): {auc_roc_knn}")

# Generate confusion matrix
cm_knn = confusion_matrix(y_test, y_pred_knn)
plt.figure(figsize=(6, 4))
sns.heatmap(cm_knn, annot=True, fmt="d", cmap="Blues")
plt.xlabel("Predicted")
plt.ylabel("True")
plt.title("Confusion Matrix (KNN)")
plt.show()

# Generate classification report
print(classification_report(y_test, y_pred_knn))

#LOGISTIC REGRESSION


# Calculate evaluation metrics
accuracy_logreg = accuracy_score(y_test, y_pred_logreg)
precision_logreg = precision_score(y_test, y_pred_logreg)
recall_logreg = recall_score(y_test, y_pred_logreg)
f1_logreg = f1_score(y_test, y_pred_logreg)
auc_roc_logreg = roc_auc_score(y_test, y_pred_logreg)

print(f"Accuracy (Logistic Regression): {accuracy_logreg}")
print(f"Precision (Logistic Regression): {precision_logreg}")
print(f"Recall (Logistic Regression): {recall_logreg}")
print(f"F1-score (Logistic Regression): {f1_logreg}")
print(f"AUC-ROC (Logistic Regression): {auc_roc_logreg}")

# Generate confusion matrix
cm_logreg = confusion_matrix(y_test, y_pred_logreg)
plt.figure(figsize=(6, 4))
sns.heatmap(cm_logreg, annot=True, fmt="d", cmap="Blues")
plt.xlabel("Predicted")
plt.ylabel("True")
plt.title("Confusion Matrix (Logistic Regression)")
plt.show()
# Generate classification report
print(classification_report(y_test, y_pred_logreg))


```
